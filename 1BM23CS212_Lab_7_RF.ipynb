```python
import pandas as pd
df=pd.read_csv("iris (1).csv")
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Separate features (X) and target (y)
X = df.drop('species', axis=1)
y = df['species']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize and train the classifier with n_estimators=10
model_default = RandomForestClassifier(n_estimators=10, random_state=42)
model_default.fit(X_train, y_train)

# Make predictions and evaluate the score
y_pred_default = model_default.predict(X_test)
score_default = accuracy_score(y_test, y_pred_default)

print(f"Prediction score with n_estimators=10: {score_default:.4f}")
```

```python
best_score = 0
best_n_estimators = 0

# Test a range of n_estimators values
for n in range(1, 101, 5): # From 1 to 100, step by 5
    model = RandomForestClassifier(n_estimators=n, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    score = accuracy_score(y_test, y_pred)

    if score > best_score:
        best_score = score
        best_n_estimators = n

print(f"\nBest prediction score found: {best_score:.4f}")
print(f"Corresponding n_estimators: {best_n_estimators}")
```

```python
from sklearn.metrics import confusion_matrix

# Re-train the model with the best n_estimators (1)
best_model = RandomForestClassifier(n_estimators=best_n_estimators, random_state=42)
best_model.fit(X_train, y_train)

# Make predictions with the best model
y_pred_best = best_model.predict(X_test)

# Calculate the confusion matrix
cm = confusion_matrix(y_test, y_pred_best)

print(f"Confusion Matrix for n_estimators={best_n_estimators}:")
print(cm)
```
