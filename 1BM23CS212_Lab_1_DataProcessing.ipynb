```python
import pandas as pd
df=pd.read_csv("Diabetes.csv")
df
```

```python
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)
```

```python
df.info()
```

```python
print("Unique values and counts for 'Gender' column:\n", df['Gender'].value_counts())
print("\nUnique values and counts for 'CLASS' column:\n", df['CLASS'].value_counts())
```

```python
df['Gender'] = df['Gender'].replace({'f': 'F'}).str.strip()
df['CLASS'] = df['CLASS'].str.strip()

print("Standardized unique values and counts for 'Gender' column:\n", df['Gender'].value_counts())
print("\nStandardized unique values and counts for 'CLASS' column:\n", df['CLASS'].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])

print("Encoded 'Gender' column unique values:", df['Gender'].unique())
print("First five rows with encoded 'Gender':\n", df.head())
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

onehot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
class_encoded = onehot_encoder.fit_transform(df[['CLASS']])

class_df = pd.DataFrame(class_encoded, columns=onehot_encoder.get_feature_names_out(['CLASS']))

df = pd.concat([df.drop('CLASS', axis=1), class_df], axis=1)

print("First five rows with encoded 'CLASS':\n", df.head())
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
exclude_cols = ['ID', 'No_Pation', 'Gender', 'CLASS_N', 'CLASS_P', 'CLASS_Y']

# Filter out excluded columns
for col in exclude_cols:
    if col in numerical_cols:
        numerical_cols.remove(col)

print("Numerical columns for outlier detection:", numerical_cols)
```

```python
outliers_count = {}

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count outliers
    col_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outliers_count[col] = len(col_outliers)

print("Number of outliers per numerical column (IQR method):")
for col, count in outliers_count.items():
    print(f"- {col}: {count} outliers")
```

```python
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap outliers
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

print("First five rows of the DataFrame after outlier capping:")
print(df.head())
```

```python
print("Numerical columns identified for scaling:", numerical_cols)
```

```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()
df_minmax_scaled = df.copy() # Create a copy to store min-max scaled data
df_minmax_scaled[numerical_cols] = min_max_scaler.fit_transform(df[numerical_cols])

print("First five rows of DataFrame after Min-Max Scaling:")
print(df_minmax_scaled[numerical_cols].head())
```

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
df_standard_scaled = df.copy() # Create a copy to store standard scaled data
df_standard_scaled[numerical_cols] = standard_scaler.fit_transform(df[numerical_cols])

print("First five rows of DataFrame after Standard Scaling:")
print(df_standard_scaled[numerical_cols].head())
```
